```python
#! Colocar como utils .py

import time
import psutil
import pandas as pd
from memory_profiler import memory_usage

def benchmark_training(model, X_train, y_train):
    """
    Mede tempo, CPU e memória durante o treinamento.
    """

    cpu_before = psutil.cpu_percent(interval=None)

    start_time = time.time()

    memory = memory_usage(
        (model.fit, (X_train, y_train)),
        interval=0.1, #Verifica a memoria a cada 100 milisegundos
        max_usage=True
    )

    end_time = time.time()

    cpu_after = psutil.cpu_percent(interval=None)

    return {
        "training_time_sec": round(end_time - start_time, 2),
        "peak_memory_mb": round(memory, 2),
        "cpu_usage_percent": cpu_after,
    }
```

```python
def compare_scenarios(raw_metrics, clean_metrics):

    return pd.DataFrame({
        "Metric": [
            "Training Time (s)",
            "Peak Memory (MB)"
        ],
        "Raw Data": [
            raw_metrics["training_time_sec"],
            raw_metrics["peak_memory_mb"]
        ],
        "Clean Data": [
            clean_metrics["training_time_sec"],
            clean_metrics["peak_memory_mb"]
        ]
    })
```
